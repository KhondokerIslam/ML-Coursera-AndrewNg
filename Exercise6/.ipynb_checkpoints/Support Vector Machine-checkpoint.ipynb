```python
import os
import numpy as np
import re
from matplotlib import pyplot
from scipy import optimize
from scipy.io import loadmat
import utils
grader = utils.Grader()
from operator import itemgetter

%matplotlib inline
```

```python
def gaussianKernel(x1,x2,sigma):
    sim = 0
    sim = np.exp(- np.sum((x1-x2)**2)/(2*(sigma**2)) )
    return sim

x1 = np.array([1,2,-1])
x2 = np.array([0, 4, -1])

sigma = 2

sim = gaussianKernel(x1,x2,sigma)
print('Gaussian Kernel between x1 = [1, 2, 1], x2 = [0, 4, -1], sigma = %0.2f:'
      '\n\t%f\n(for sigma = 2, this value should be about 0.324652)\n' % (sigma, sim))
```

```python
grader[1] = gaussianKernel
grader.grade()
```

```python
data = loadmat(os.path.join('Data', 'ex6data3.mat'))
X, y, Xval, yval = data['X'], data['y'][:, 0], data['Xval'], data['yval'][:, 0]

utils.plotData(X, y)
```

```python
def dataset3Params(X, y, Xval, yval):
    C = 1
    sigma = 0.3
    errorRow = 0
    C_text = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])
    sigmaTest = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])
    results = np.eye(64,3)
    
    for c in C_text:
        for s in sigmaTest:
            model = utils.svmTrain(X, y, c, gaussianKernel, args = (s, ))
            predictions = utils.svmPredict(model, Xval)
            prediction_error = np.mean(predictions !=yval )
        
            results[errorRow:] = [c, s, prediction_error]
    
    sorted_results = sorted(results, key=itemgetter(2))
    #print(sorted_results.shape)
    C = sorted_results[0][0]
    S = sorted_results[0][1]
    return C, S

C, sigma = dataset3Params(X, y, Xval, yval)

# Train the SVM
# model = utils.svmTrain(X, y, C, lambda x1, x2: gaussianKernel(x1, x2, sigma))
model = utils.svmTrain(X, y, C, gaussianKernel, args=(sigma,))
utils.visualizeBoundary(X, y, model)
print(C, sigma)
```

```python
grader[2] = lambda : (C, sigma)
grader.grade()
```
